# the last vehicle (1026) has a time offset in on the the loggers, but what is the offset? the data still looks good, just someone messed up local versus GMT or something?
- below here we are going to figure out the time offset by starting with hours, then minutes, then seconds.

```python
#find hour with minimum difference by hour
ds_sbe2 = ds_sbe.copy(deep=True)
for ih in range(-30,-20):
    ds_sbe2['time']=ds_sbe['time']+np.timedelta64(ih,'h')
    dif = np.abs((ds_sbe.temp[2,2,:]-ds_sbe2.temp[2,3,:])).mean()
    print(ih,dif.data)
```

```python
#find minute with minimum difference
for ih in range(-8,8):
    deltaT=-28*60+ih   #minimum at 28 hours
    ds_sbe2['time']=ds_sbe['time']+np.timedelta64(deltaT,'m')
    dif = np.abs((ds_sbe.temp[2,2,:]-ds_sbe2.temp[2,3,:])).mean()
    print(ih,dif.data)
```

```python
#find minute with minimum difference
for ih in range(-25,-15):
    deltaT=-28*60*60+4*60+ih
    ds_sbe2['time']=ds_sbe['time']+np.timedelta64(deltaT,'s')
    dif = np.abs((ds_sbe.temp[2,2,:]-ds_sbe2.temp[2,3,:])).mean()
    print(deltaT,dif.data)
```
